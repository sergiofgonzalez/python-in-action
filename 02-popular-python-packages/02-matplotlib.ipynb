> Python visualizations

*Matplotlib* is a comprehensive library for creating static, animated, and interactive visualizations in Python.

In this notebook you will find a set of recipes for different related *Matplotlib* charts and visualizations with different degree of difficulty.

You will typically start by importing `pyplot` as follows:

```python
import matplotlib.pyplot as plt
```

| NOTE: |
| :-------- |
| The syntax `import <module>.<sub_module>` is used to import individual objects/submodules from a package. |

#### `plt.plot(...)`: Plotting functions

Use `plt.plot(...)` to plot in 2D a mathematical function. It requires as first argument a list with the values for the *x-axis*, and as second argument, the corresponding *y-axis* values.

The following example plots `sin(x)` for x in range $[0, 4 \pi] $.

| NOTE: |
| :---- |
| When using Jupyter notebooks, invoking `plt.plot()` will display the chart. Otherwise, you will need to invoke `plt.show()` to display it. |

````python
import matplotlib.pyplot as plt
import numpy as np
from math import pi

x = np.linspace(0, 4 * pi)
plt.plot(x, np.sin(x))

plt.show() # not needed in Jupyter
````
